 ## Task 02- Performed a data cleaning and exploratory data analysis on a dataset

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df2= pd.read_csv("test.csv")
df2.describe()
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2['Age']=df2['Age'].fillna(df2['Age']).mean()
df2['Fare']=df2['Fare'].fillna(df2['Fare']).mean()
df2['Cabin']=df2['Cabin'].fillna('No Cabin')

df2.isnull().sum()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Sex', data=df2)
plt.title('Count of passengers by gender')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Pclass', data=df2)
plt.title('Passengers class Distributiom')
plt.xlabel('Pclass')
plt.ylabel('Count')

plt.show()
```

```python
df3= pd.read_csv("train.csv")

new_column= df3['Survived']
df2['Survived']= new_column
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='Sex', y='Survived', data=df2)
plt.title('Survival rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
df2.head()

df2['Title'] = df2['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
plt.figure(figsize=(10, 6))
sns.countplot(x='Title', data=df2)
plt.title('Count of Passengers by Title')
plt.xlabel('Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
